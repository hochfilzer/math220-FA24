# Plotting and visualizing vectors and subspaces (spans)


```python
import numpy as np
import matplotlib.pyplot as plt
```

Let's start with something simple. Say you would like to plot a vector in $\mathbb{R}^2$, for example the vectors $\overrightarrow{u} = \begin{pmatrix} 1 \\2 \end{pmatrix}$, $\overrightarrow{v} = \begin{pmatrix} 3 \\-4 \end{pmatrix}$ and $\overrightarrow{w} = \begin{pmatrix} -1 \\4 \end{pmatrix}$. We use matplotlib to first initialize a 2D plane and then with a little bit of fiddling around we can represent a vector with a fat line and an arrow at its head.

```python
fig, ax = plt.subplots(figsize = (7,7)) # Initialize a coordinate system and how big we want it to be (on our screen)
u = np.array([1,2])
v = np.array([3,-4])
w = np.array([-1,4])

vectors = [u,v,w]
colors = ['red', 'blue', 'green']
for vector, c in zip(vectors, colors):
    ax.arrow(0, 0, vector[0], vector[1], color=c, width=0.08, 
             length_includes_head=True, head_width=0.3, head_length=0.4)


ax.grid(True)
ax.axis([-5,5,-5,5]) # Here we tell python the range of the axis that we want to see
ax.set_xlabel('x-axis', size = 18)
ax.set_ylabel('y-axis', size = 18)

plt.show()



```

Next we may want to visualize the span of a set of vectors in two dimensions. There isn't very much that can happen here. Either we get the zero vector (infinitely little to visualize here),  we get the whole of $\mathbb{R}^2$ which again is just essentially plotting a coordinate system or finally and more interestingly we could get a line passing through the origin, which we can do as follows:

Given the vectors $\overrightarrow{u},$  $\overrightarrow{v},$  $\overrightarrow{w}$ we will now plot the lines spanned by these vectors respectively.

```python
fig, ax = plt.subplots(figsize = (7,7)) # Initialize a coordinate system and how big we want it to be (on our screen)

x = np.linspace(-5, 5, 1000) # First we create 1,000 uniformly distributed points between -5 and 5 which we will use to create corresponding 
# and y values so that we can plot those points as a line

for vector, c in zip(vectors, colors):
    plt.plot(x * vector[0], x * vector[1], color = c, lw = 3, label = f"Subspace spanned by {vector}")

ax.grid(True)
ax.axis([-5,5,-5,5]) # Here we tell python the range of the axis that we want to see

ax.set_xlabel('x-axis', size = 18)
ax.set_ylabel('y-axis', size = 18)
plt.legend() # This will display the legend

plt.show()

```

If you want to plot the span of a set of vectors in two dimensions you can do so via the following:

```python
fig, ax = plt.subplots(figsize = (8,8))


s = np.linspace(-5, 5, 100)
t = np.linspace(-5, 5, 100)
S, T = np.meshgrid(s, t)

X = u[0] * S + v[0] * T
Y = u[1] * S + v[1] * T

plt.scatter(X, Y)
ax.axis([-5,5,-5,5]) # Here we tell python the range of the axis that we want to see


ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')


plt.show()
```

The above plot should essentially tell you that the whole of $\mathbb{R}^2$ is spanned by the vectors $\overrightarrow{u}$ and $\overrightarrow{v}$. Finally, let's see what happens if we plot the span of the two vectors $\overrightarrow{u}$ and $3\overrightarrow{u}$.

```python
fig, ax = plt.subplots(figsize = (8,8))


s = np.linspace(-5, 5, 100)
t = np.linspace(-5, 5, 100)
S, T = np.meshgrid(s, t)

X = u[0] * S + 3 * u[0] * T
Y = u[1] * S + 3 * u[1] * T

plt.scatter(X, Y)
ax.axis([-5,5,-5,5]) # Here we tell python the range of the axis that we want to see


ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')


plt.show()
```

We find that we just get the line spanned by $\overrightarrow{u}$.

Finally we are also interested vectors and subspaces in three dimensions. Let's start by visualizing the vectors $\overrightarrow{a} = \begin{pmatrix} 1 \\2 \\ 0 \end{pmatrix}$, $\overrightarrow{b} = \begin{pmatrix} 3 \\-4 \\ 9 \end{pmatrix}$ and $\overrightarrow{c} = \begin{pmatrix} -1 \\4 \\ -4 \end{pmatrix}$.


```python
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(1, 1, 1, projection='3d')

a = np.array([1,2,0])
b = np.array([3,-4,9])
c = np.array([-1, 4, -4])


ax.quiver(0,0,0,a[0], a[1], a[2], color = 'red')
ax.quiver(0,0,0,b[0], b[1], b[2], color = 'blue')
ax.quiver(0,0,0,c[0], c[1], c[2], color = 'green')


ax.set_xlim([-3,3])
ax.set_ylim([-3,3])
ax.set_zlim([-3,3])

ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z-axis')
ax.set_box_aspect(None, zoom=0.85) # a workaround needed to display the z-axis label

plt.show()

```

Here is how to plot the lines spanned by the vectors that considered above.

```python
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(1, 1, 1, projection='3d')

x = np.linspace(-3,3,1000)

a = np.array([1,2,0])
b = np.array([3,-4,9])
c = np.array([-1, 4, -4])


ax.plot(a[0] * x, a[1] * x, a[2] * x, color = 'red', label = 'Line spanned by a')
ax.plot(b[0] * x, b[1] * x, b[2] * x, color = 'blue', label = 'Line spanned by b')
ax.plot(c[0] * x, c[1] * x, c[2] * x, color = 'green', label = 'Line spanned by c')



ax.set_xlim([-3,3])
ax.set_ylim([-3,3])
ax.set_zlim([-3,3])

ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z-axis')
ax.set_box_aspect(None, zoom=0.85) # a workaround needed to display the z-axis label

plt.show()
```

Finally, let's plot the plane spanned by $\overrightarrow{a}$ and $\overrightarrow{b}$.

```python
fig = plt.figure(figsize = (8,8))

ax = fig.add_subplot(111,projection='3d')

s = np.linspace(-5, 5, 10)
t = np.linspace(-5, 5, 10)
S, T = np.meshgrid(s, t)

X = a[0] * S + b[0] * T
Y = a[1] * S + b[1] * T
Z = a[2] * S + b[2] * T

ax.plot_surface(X, Y, Z, alpha = .9)


ax.set_title('Visualization of a plane in R3')

ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z-axis')
ax.set_box_aspect(None, zoom=0.85) # a workaround needed to display the z-axis label


plt.show()
```
